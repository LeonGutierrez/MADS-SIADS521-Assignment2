<h2><span style="color:green">What is Python Seaborn library?</span></h2>

Seaborn is a tabular/declarative plotting library. Plotting libraries like Seaborn, can treat columns of data frames as properties of a graph at a high level. Seaborn is a python library that is used to create graphics in Python. It is built on top of matplotlib and integrated with pandas, scypy, and numpy. This package is used to simplify complex data visualizations. 
Some of the features Seaborn offers are:
<ul>
    <li>An API that is dataset-oriented to allow comparison between multiple variables</li>
    <li>Supports multi-plot grids allowing the user to build complex visualizations</li>
    <li>Univariate and bivariate charts to compare between subsets of data</li>
    <li>Availability of different color palettes to better highlight variable patterns</li>
    <li>Estimates and plots linear regression automatically for different types of dependent vaiables</li>
    <li>Availability of different color palettes to better highlight variable patterns</li>
</ul>

Seaborn aims to make visualization a central part of exploring and understanding data. Its dataset-oriented plotting functions operate on dataframes and arrays containing whole datasets and internally perform the necessary semantic mapping and statistical aggregation to produce informative plots.

<strong>In summary, I chose Seaborn for this tutorial because Seaborn is very simple to use compare to other packages like Matplotlib. I makes hard visualizations easy to to be defined. I personally like  that it automatically displays linear regressions when I analyze scatterplots</strong><br>

<h2><span style="color:green">Why using scatterplots to analyze bivariate relationships?</span></h2>

In general, I think scatterplot are a great way to explore your data when you want to find relationships in your data. That relationship can be calculated as a scalar (Pearson or Spearman). However, a visual exploration provides insights not capture by a single number. This exploratory analysis is very useful in cases where you want to predict a value, so one variable's final estimation will depend on the value of another one. In summary, this types of analysis are good in cases:
1. When you want to analize paired numerical data. 
2. When trying to determine whether the variables you are analyzing are correlated. This correlation helps determine whether one variable could be used to explain the behavior of the other
3. In classification and regression problems, scatterplots allow you to see these correlations in a way that it is easy to determine explanatory power

<strong>From the above scatterplot</strong>: Although it is possible to see that there is some positive relationship of this two variables, we probably want to explore if other variables might add more insight to this two variables relationship.
This tutorial does not explore all Seaborn aesthetic options, however you can run <code>sns.axes_style()</code> in order to see all defaul options. Then you can make changes accordingly.

```python
###################################################
#### This variation provides a little       #######
#### twist to the a simple scatter plot     #######
#### It is very simple to define and you    #######
#### are able to display a regression       #######
#### line and see variable distributions    #######
###################################################
sns.jointplot(x="alcohol", 
              y="pH", 
              data=reds,
              kind="reg",
              line_kws={'color': 'black'})
```

<strong>From the above plot</strong>: This variation from the first 'vanilla' scatterplot adds a view that allows is to see the distribution of each variable individually. Sometimes it might be useful to see such distributions as a way to spot potential anomalies in the data or potential ranges to analize.  

```python
###################################################
#### With this variation we want to have a  #######
#### better grasp of the same 2 variables   #######
#### but with a secondary variable as the   #######
#### colour encoding variable.              #######
###################################################
pal = sns.dark_palette("palegreen", as_cmap=True)

sns.set_context("paper", rc={"font.size":12,"axes.titlesize":12,"axes.labelsize":12}) 
sns.lmplot(x="alcohol", 
            y="pH",
            hue="Wine Quality",                                   # This option controls the encoding for 
            data=reds,
            markers=[".","+"]).set(title = 'Bivariate Analysis',  # Adding additional control over the chart 
                            xlabel = 'Free SO2 Content', 
                            ylabel = 'Alcohol Content' );
 

```

<strong>From the above scatterplot:</strong> By adding a hue variable, it is possible to predict or have a sense of variable separation potential. In this case we are trying to determine whether these two variables have the separation potential for the variable <code>Wine Quality</code>. 

```python
###################################################
#### With this variation we are able to     #######
#### create a grid with Wine Quality and    #######
#### Acidity to determine. The advantage    #######
#### of this approach is that effects are   #######
#### separated for better visualization     #######
###################################################
sns.set_style("darkgrid", {"axes.facecolor": ".85"})
sns.lmplot(x="free sulfur dioxide", 
           y="alcohol",
           col="Wine Quality",               # 'col=' and 'row=' Variables that define subsets of the data, 
           row="Acidity",                    # which will be drawn on separate facets in the grid
           data=reds)
```

<strong>From the above plots grid:</strong> This option allows to plot data and regression model fits across a FacetGrid. This new perspective is able to separate interacions that are easily spotted with the regression line.
